```python
from bs4 import BeautifulSoup
import requests
```

```python
url_drivers = 'https://www.formula1.com/en/results/2024/drivers'
```

```python
page = requests.get(url_drivers)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
drivers = soup.find('table', class_ = 'f1-table f1-table-with-data w-full')
print(drivers)
```

```python
drivers_columns = drivers.find_all('th')
print(drivers_columns)
```

```python
column_titles = [th.find('p').get_text(strip=True) for th in drivers_columns]
print(column_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = column_titles)

df
```

```python
column_data = drivers.find_all('tr')[1:]
print(column_data[0])
```

```python
row_data = []

for row in column_data:
    p_tags = row.find_all('p')
    row_text = [p.get_text(strip=True) for p in p_tags]
    row_data.append(row_text)

for data in row_data:
    print(data)
```

```python
for row in row_data:
    if len(row) >= 2:
        row[1] = row[1][:-3]

for data in row_data:
    print(data)
```

```python
for row in row_data:
    df.loc[len(df)] = row

df
```

```python
df.to_csv('drivers.csv', index = False)
```
